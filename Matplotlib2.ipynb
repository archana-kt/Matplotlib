```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(100, size = 100)
```

```python
x
```

```python
x = np.linspace(10,20,100)
```

```python
x
```

```python
plt.plot(x,np.sin(x))
```

```python
# Using limits in x and y axis
```

```python
plt.plot(x,np.sin(x))
plt.xlim(10,14)
plt.ylim(-0.5,+0.5)
```

```python
# Also we can give axis instead of xlim and ylim
```

```python
plt.plot(x,np.sin(x))
plt.axis([12,16,-0.5,+0.5])
```

```python
plt.plot(x,np.sin(x))
plt.axis("tight") # it will tightening the axis values
```

```python
plt.plot(x,np.sin(x))
plt.axis("equal") # it will tightening the axis values
```

```python
# Using Legend - for creating label
```

```python
plt.plot(x,np.sin(x),label="Sin")
plt.plot(x,np.cos(x),label="Cos")
```

```python
# See here there is no label is showing, for that we can use legend
```

```python
plt.plot(x,np.sin(x),label="Sin")
plt.plot(x,np.cos(x),label="Cos")
plt.legend()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') # g=green
plt.plot(x, np.cos(x), ':b', label='cos(x)') # b=blue
plt.axis('equal')
plt.legend()
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black');
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o-', color='black');
```

```python
plt.plot(x, y, '-p', color='gray', markersize=15, linewidth=4, markerfacecolor='white', markeredgecolor='gray',markeredgewidth=2)
```

```python
plt.plot(x, y, '-p', color='blue', markersize=15, linewidth=4, markerfacecolor='white', markeredgecolor='red',markeredgewidth=2)
```
